```python
import numpy as np

# Create the 2D array
matrix = np.array([[18, 21, 3, 6],
                   [5, 1, 0, 11],
                   [0, 7, 0, 33]])

# a. Print indices of 0 (zero) values
zero_indices = np.argwhere(matrix == 0)
print("Indices of 0 values:\n", zero_indices)

# b. Find transpose of the matrix and extract 2nd row
transposed_matrix = matrix.T
second_row_transposed = transposed_matrix[1]
print("Transpose of the matrix:\n", transposed_matrix)
print("2nd row of transposed matrix:", second_row_transposed)

# c. Flatten the original matrix
Z = matrix.flatten()
print("Flattened matrix:", Z)

# d. Write original matrix to 'mat.txt' file
np.savetxt('mat.txt', matrix, fmt='%d')
print("Matrix written to 'mat.txt' file.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create and save the marks data to 'mark.txt'
marks_data = np.array([[60, 66, 70],
                       [99, 78, 88],
                       [55, 61, 54],
                       [84, 81, 73],
                       [55, 54, 53],
                       [77, 72, 79],
                       [80, 80, 80],
                       [90, 95, 93]])
np.savetxt('mark.txt', marks_data, fmt='%d')

# Load the data from 'mark.txt'
marks = np.loadtxt('mark.txt', dtype=int)

# a. Print max of mark1, mark2, and mark3
max_mark1 = np.max(marks[:, 0])
max_mark2 = np.max(marks[:, 1])
max_mark3 = np.max(marks[:, 2])

print(f"Maximum Marks:")
print(f"Mark1: {max_mark1}")
print(f"Mark2: {max_mark2}")
print(f"Mark3: {max_mark3}")

# b. Plot scatterplots
plt.figure(figsize=(10, 5))

# Scatterplot for mark1 vs mark2
plt.subplot(1, 2, 1)
plt.scatter(marks[:, 0], marks[:, 1], color='b')
plt.title("Mark1 vs Mark2")
plt.xlabel("Mark1")
plt.ylabel("Mark2")

# Scatterplot for mark1 vs mark3
plt.subplot(1, 2, 2)
plt.scatter(marks[:, 0], marks[:, 2], color='r')
plt.title("Mark1 vs Mark3")
plt.xlabel("Mark1")
plt.ylabel("Mark3")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Create the initial DataFrame
data = pd.DataFrame({
    'A': [1, 4, 7],
    'B': [2, 5, 8],
    'C': [3, 6, 9]
})

# a. Add 5 rows to the DataFrame
new_rows = pd.DataFrame({
    'A': [10, 13, 16, 19, 22],
    'B': [11, 14, 17, 20, 23],
    'C': [12, 15, 18, 21, 24]
})
data = pd.concat([data, new_rows], ignore_index=True)
print("DataFrame after adding 5 rows:")
print(data, "\n")

# b. Print rows from 1 to 4 from beginning and end
print("Rows from 1 to 4 from beginning (using head):")
print(data.head(4), "\n")

print("Rows from 1 to 4 from end (using tail):")
print(data.tail(4), "\n")

# c. Add a new column
data['D'] = [25, 26, 27, 28, 29, 30, 31, 32]
print("DataFrame after adding a new column:")
print(data, "\n")

# d. Drop the first column
data_dropped = data.drop('A', axis=1)
print("DataFrame after dropping the first column:")
print(data_dropped)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the iris.csv file into a DataFrame
iris_df = pd.read_csv('Iris.csv')

# a. Print names of all features and the shape of the dataset
feature_names = iris_df.columns.tolist()
dataset_shape = iris_df.shape
print("Feature Names:", feature_names)
print("Dataset Shape:", dataset_shape)

# b. Find descriptive statistics
descriptive_stats = iris_df.describe()
print("\nDescriptive Statistics:")
print(descriptive_stats)

# c. Count distinct species and samples in each class
species_counts = iris_df['Species'].value_counts()
print("\nSpecies Counts:")
print(species_counts)

# d. Extract rows for each species
df_setosa = iris_df[iris_df['Species'] == 'Iris-setosa']
df_versicolor = iris_df[iris_df['Species'] == 'Iris-versicolor']
df_virginica = iris_df[iris_df['Species'] == 'Iris-virginica']

# e. Visualize the correlation between SepalLengthCm and SepalWidthCm
plt.figure(figsize=(8, 6))
plt.scatter(iris_df['SepalLengthCm'], iris_df['SepalWidthCm'], c='b', alpha=0.6)
plt.title("Correlation between Sepal Length and Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
